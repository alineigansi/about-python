```
#Função len() conta os espaços também
len('Suave na nave') 
```

```
#Tanto faz usar print() ou apenas chamar diretamente pelo nome da variável (no caso de Strings aparecem as aspas)
s = 'Suave na nave'
s
```

```
print(s)
```

```
#Indexação começa em zero
#Mostra o primeiro elemento
s[0]
```

```
#Retorna todos os elementos a partir do elemento de índice 1
s[1:]
```

```
#Retorna tudo até o elemento de índice 3 - pega tudo de 0 a 3, não inclui o elemento 3!
s[:3]
```

```
#Tudo
s[:]
```

```
#Indexação negativa para retroceder
s[:-1]
```

```
#Última letra (dois índices antes do 0, então ele começa da parte de trás)
s[-2]
```

```
#Pega tudo, de 2 em 2
s[::2]
```

```
#Pega tudo, mas com passos negativos, de trás para frente
s[::-1]
```
